```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
device
```

```python
device = torch.device("cpu")
```

```python
device
```

```python
y_train
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

y_train_tensor, y_test_tensor = torch.tensor(y_train, dtype=torch.long), torch.tensor(y_test, dtype=torch.long)
```

```python
X_train_tensor.shape
```

```python
y_train_tensor
```

```python
X_train_tensor = X_train_tensor.unsqueeze(1)
X_test_tensor = X_test_tensor.unsqueeze(1)
```

```python
X_train_tensor.shape
```

```python
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)
```

```python
class EmotionCNN(nn.Module):
    def __init__(self, num_classes=8):
        super().__init__()

        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.fc_layers = nn.Sequential(
            nn.Linear(128 * 4 * 37, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )
    def forward(self, x):
        x = self.conv_layers(x)
        x = torch.flatten(x, 1)
        x = self.fc_layers(x)

        return x
```

```python
emotion_cnn_model = EmotionCNN(num_classes=8).to(device)
```

```python
cross_loss = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(emotion_cnn_model.parameters(), lr=0.001)
```

```python
# Put the target data on device
X_train_tensor = X_train_tensor.to(device)
X_test_tensor = X_test_tensor.to(device)
y_train_tensor, y_test_tensor = y_train_tensor.to(device), y_test_tensor.to(device)
```

```python
X_train_tensor.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Evalution of Model

train_loss_value_cnn = []
test_loss_value_cnn = []
train_accuracy_cnn = []
test_accuracy_cnn = []

epoch_counts = []

torch.manual_seed(42)

epochs = 30

for epoch in tqdm(range(epochs), desc="Training CNN Model"):

    # Training phase
    emotion_cnn_model.train()

    total_train_loss = 0
    total_train_correct = 0
    total_train_samples = 0

    # Forward pass
    for X_batch, y_batch in tqdm(train_loader, desc="Training Batches", leave=False):

        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        y_train_logits = emotion_cnn_model(X_batch)
        y_train_pred = torch.argmax(y_train_logits, dim=1)

        # Calculate loss
        train_loss = cross_loss(y_train_logits, y_batch)

        # Calculate accuracy
        train_acc = accuracy_score(y_batch.cpu(), y_train_pred.cpu())

        # Optimzer zero grad
        optimizer.zero_grad()

        # Loss Backward
        train_loss.backward()

        # Optimizer step
        optimizer.step()

        # Accumulate loss
        total_train_loss  += train_loss.item() * X_batch.size(0)
        total_train_correct += (y_train_pred == y_batch).sum().item()
        total_train_samples += X_batch.size(0)

    # Model testing phase

    emotion_cnn_model.eval()
    total_test_loss = 0
    total_test_correct = 0
    total_test_samples = 0

    with torch.inference_mode():
        for X_batch_test, y_batch_test in tqdm(test_loader, desc="Testing Batches" ,leave=False):
            X_batch_test, y_batch_test = X_batch_test.to(device), y_batch_test.to(device)
            test_logits = emotion_cnn_model(X_batch_test)
            test_pred = torch.argmax(test_logits, dim=1)

            # Calculate loss
            test_loss = cross_loss(test_logits, y_batch_test)
            # Calculate accuracy
            test_acc = accuracy_score(y_batch_test.cpu(), test_pred.cpu())

            # Accumulate loss
            total_test_loss += test_loss.item() * X_batch_test.size(0)
            total_test_correct += (test_pred == y_batch_test).sum().item()
            total_test_samples += X_batch_test.size(0)
    
    # Calculate average losses and accuracies
    epoch_train_loss = total_train_loss / total_train_samples
    epoch_test_loss = total_test_loss / total_test_samples

    epoch_train_acc = total_train_correct / total_train_samples * 100
    epoch_test_acc = total_test_correct / total_test_samples * 100

    if epoch % 5 == 0:
        train_loss_value_cnn.append(epoch_train_loss)
        test_loss_value_cnn.append(epoch_test_loss)
        train_accuracy_cnn.append(epoch_train_acc)
        test_accuracy_cnn.append(epoch_test_acc)
        epoch_counts.append(epoch)

        print(f"Epoch: {epoch} | "
              f"Train Loss: {epoch_train_loss:3f} | Train Accuracy: {epoch_train_acc:.2f}% |"
              f"Test Loss: {epoch_test_loss:.3f} | Test Accuracy: {epoch_test_acc:.2f}")
```
